```python
df[(df['Destination'] == '192.168.0.4') & (df['Source'] == '10.0.1.6')]
```

```python
df_sus = df[((df['Destination'] == '192.168.0.4') & (df['Source'] == '10.0.1.6')) | ((df['Source'] == '192.168.0.4') & (df['Destination'] == '10.0.1.6'))]
df_sus
```

```python
df_sus['Src port'] = df_sus['Info'].str.extract(r'(^\d*)')
df_hist = df_sus.groupby('Src port').Source.count()
df_hist
```

```python
%matplotlib
```

```python
df_hist2 = df_hist.reset_index()
df_hist3 = df_hist2.rename(columns={'Source': 'Count'})
df_hist3
```

```python
df_hist4 = df_hist3[df_hist3['Count'] > 5]
df_hist4
```

```python
df_hist5 = df_hist4.drop(0)
df_hist5
```

```python
df_hist5.plot(x='Src port', y='Count', kind='bar')
```
